```python
data = np.load('kdd_cup.npz')

y = data["kdd"][:,-1]
X = data["kdd"][:,:-1]
```

```python
y0 = y[y==0].shape[0]
y1 = y[y==1].shape[0]
print(y0, y1)

labels = ['normal', 'anomaly']
y_percent = [y0/(y0+y1)*100, y1/(y0+y1)*100]

plt.pie([y0, y1], labels=labels, autopct='%1.1f%%')
```

```python
X.shape
```

```python
from sklearn.decomposition import PCA

pca = PCA().fit(X)
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel("number of components")
plt.ylabel("cumulative explained variance")
plt.grid(linestyle=":")
plt.show()

pca = PCA(0.90).fit(X)
print(f"N components: {pca.n_components_}")
```

```python
X = PCA(0.90).fit_transform(X)
X.shape
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42, stratify=y)
```

```python
print(f"X_train - y_train dim: {X_train.shape} - {y_train.shape}")
print(f"X_test - y_test dim: {X_test.shape} - {y_test.shape}")

print(f"class ratio on train (0 vs. 1): {y_train[y_train==0].shape[0]} vs. {y_train[y_train==1].shape[0]}")
print(f"class ratio on test (0 vs. 1): {y_test[y_test==0].shape[0]} vs. {y_test[y_test==1].shape[0]}")
```

```python
from sklearn import svm

clf = svm.SVC(kernel="rbf").fit(X_train, y_train)
y_svm_pred = clf.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score

acc_svm = accuracy_score(y_test, y_svm_pred)
acc_svm
```

```python
from sklearn.metrics import roc_curve, auc


fpr_svm, tpr_svm, _ = roc_curve(y_test, y_svm_pred)
roc_auc_svm = auc(fpr_svm, tpr_svm)
roc_auc_svm
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,10))

plt.title('Receiver Operating Characteristic')

plt.plot(fpr_svm, tpr_svm, color='blue',label = f'SVM (AUC = {roc_auc_svm:0.2f}, Acc = {acc_svm:0.2f})')

plt.legend(loc = 'lower right')
plt.plot([0, 1], [0, 1],linestyle='--')
plt.axis('tight')
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')
plt.grid(":")
plt.show()
```

```python
from sklearn.metrics import classification_report, ConfusionMatrixDisplay
from sklearn.metrics import confusion_matrix


cm = confusion_matrix(y_test, y_svm_pred, labels=np.array([0, 1]))
tn, fn, tp, fp = cm[0][0], cm[1][0], cm[1][1], cm[0][1]
print(tn, fn, tp, fp)

# cm
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=np.array([0, 1]))
disp.plot()
# plt.grid()
plt.title(f"SVM ( False Positive Rate = {fp/(fp + tn)})")
plt.show()
```
